This project is meant to analyze what neighborhood is the best in Pittsburgh using data from the WPRDC. We approached this as what neighborhood would be the most fun to live in. For this particular notebook, the metric I chose was the amount of playgrounds in each neighborhood. Here is the data.

```python
import pandas as pd

playgrounds = pd.read_csv("ground.csv")
neighborhood_rank = (
    playgrounds["neighborhood"]
    .value_counts()
    .reset_index()
)
neighborhood_rank.columns = ["Neighborhood", "Playground_Count"]
print(neighborhood_rank.head(10))
```

The best neighborhood from this metric is Squirrel Hill South because it has by far the most amount of playgrounds with eight. Using this metric, it is the best neighborhood because it has the most playgrounds for children and adults alike, with enough to ensure that there is always an open playground nearby for all residents of the neighborhood. The following neighborhoods also contain a large amount of playgrounds, but not nearly as many as Squirrel Hill South. Below is a graph of playground count in the neighborhoods of Pittsburgh. 

```python
neighborhood_rank.head(10).plot(
    kind="barh",
    x="Neighborhood",
    y="Playground_Count",
    title="Top 10 Pittsburgh Neighborhoods by Number of Playgrounds"
)
```

The conclusions we can make from this data is that most neighborhoods have a sufficient number of parks, with the majority of neighborhoods hovering between 2-5, but Squirrell Hill South has a significantly larger amount of parks making it one of the best neighborhoods in Pittsburgh for fun.
